```python
import pandas as pd

# Tên tệp sẽ lấy từ uploaded
file_name = "DATA.csv"  # Đổi tên nếu cần thiết

# Đọc tệp CSV vào DataFrame
df1 = pd.read_csv(file_name)

df1
```

```python
df1['product_link'] = df1['product_id'].apply(lambda x: f"https://www.amazon.com/dp/{x}")
```

```python
# Chuyển tên cột title -> product_name và price -> actual_price
df1.rename(columns={'title': 'product_name', 'price': 'discounted_price', 'reviews': 'rating_count'}, inplace=True)
```

```python
def convert_rating(rating):
    try:
        return float(rating.split(' ')[0]) if isinstance(rating, str) and 'out of' in rating else None
    except Exception as e:
        return None

df1['rating'] = df1['rating'].apply(convert_rating)
```

```python
def convert_rating_count(rating_count):
    try:
        return int(rating_count.split(' ')[0].replace(',', '')) if isinstance(rating_count, str) else None
    except Exception as e:
        return None

df1['rating_count'] = df1['rating_count'].apply(convert_rating_count)
```

```python
import re

def handle_discount_percentage(discount):
    if isinstance(discount, str):
        if "No discount" in discount:
            return "0%"
        match = re.search(r'(-?\d+)', discount)
        if match:
            value = int(match.group(1))
            return f"{abs(value)}%"
    return "0%"

df1['discount_percentage'] = df1['discount_percentage'].apply(handle_discount_percentage)
```

```python
df1['availability'] = df1['availability'].fillna('Out of stock')
```

```python
df1['actual_price'] = df1['actual_price'].astype(str).apply(lambda x: x.replace('$', '') if isinstance(x, str) else x)

def remove_extra_dot(value):
    if isinstance(value, str):
        return value.replace('..', '.')
    return value

df1['discounted_price'] = df1['discounted_price'].apply(remove_extra_dot)
df1['actual_price'] = df1['actual_price'].apply(remove_extra_dot)
```

```python
df1['product_name'] = df1['product_name'].str.replace('/', '_')
df1['product_name'] = df1['product_name'].str.replace(r'\s+', '_', regex=True)

df1['product_name'] = df1['product_name'].str.replace(r'[^\w\s]', '', regex=True)
print(df1.head())
```

```python
def process_commas(value):
    try:
        if isinstance(value, str):
            return float(value.replace(',', ''))
        return value
    except ValueError:
        return None

df1['discounted_price'] = df1['discounted_price'].apply(process_commas)
df1['actual_price'] = df1['actual_price'].apply(process_commas)

print(df1[['discounted_price', 'actual_price']])
```

```python
# Sắp xếp lại thứ tự các cột
df1 = df1[['brand', 'product_id', 'product_name', 'type_product', 'discounted_price', 'discount_percentage',
           'actual_price', 'rating' ,'rating_count', 'about_product', 'product_link']]
```

```python
print("Dữ liệu ban đầu:")
print(df1.head())

initial_count = len(df1)

df1_cleaned = df1[~df1.isin(['Error']).any(axis=1)]

df1_cleaned = df1_cleaned.dropna()

df1_cleaned = df1_cleaned.drop_duplicates()

df1_cleaned['about_product'] = df1_cleaned['about_product'].apply(lambda x: str(x).replace("'", "").replace('"', "").replace("{", "").replace("}", "") if isinstance(x, str) else x)

final_count = len(df1_cleaned)

deleted_rows = initial_count - final_count
print(f"Số lượng dòng dữ liệu bị xoá: {deleted_rows}")
```
